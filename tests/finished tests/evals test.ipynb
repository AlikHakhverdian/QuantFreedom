```python
import pandas as pd
import numpy as np
from quantfreedom.utils import clear_cache
from quantfreedom.backtester.indicators.talib_ind import from_talib
from quantfreedom.backtester.evaluators.evaluators import eval_is_below, eval_is_above, combine_evals
prices = pd.read_csv('30min.csv', index_col='time')
```

```python
rsi_ind = from_talib(
    func_name='rsi',
    df_prices=prices,
    cart_product=False,
    combos=False,
    timeperiod=[15, 30],
)
rsi_ind
```

```python
rsi_eval = eval_is_below(
    rsi_ind,
    np.arange(30, 51, 10),
)
```

```python
ema_ind = from_talib(
    func_name='ema',
    df_prices=prices,
    cart_product=False,
    combos=False,
    timeperiod=[200, 400, 600]
)
```

```python
ema_eval = eval_is_above(
    ind_data=ema_ind,
    df_prices=prices,
    cand_ohlc='close',
)
```

```python
combine_evals(
    bigger_df=rsi_eval,
    smaller_df=ema_eval,
)
```
